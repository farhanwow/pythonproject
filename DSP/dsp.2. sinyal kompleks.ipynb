```
#Klasifikasi sinyal
#sinyal nyata dan kompleks
import numpy as np
import matplotlib.pyplot as plt

# Parameter
a = -5 # rentang waktu
b = 5
dt = 0.1
sigma = 0  # s = 0 +0j
omega = 0

# PROSEDUR
t = np.arange(a, b, dt)
s = complex(sigma,omega)
sinyal = np.exp(s*t)

# OUTPUT - grafik
plt.figure(1)
plt.plot(t,np.real(sinyal),label='real')
plt.plot(t,np.imag(sinyal),label='imagijiner')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('Sinyal  e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
```

```
# eksponensial sederhana σ> 0 ω = 0 dimana s = σ + jω 
sigma = 1
omega = 0

s = complex(sigma,omega)
sinyal = np.exp(s*t)

# Keluaran grafik
plt.figure(2)
plt.plot(t,np.real(sinyal),label='real')
plt.plot(t,np.imag(sinyal),label='imajiner')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('sinyal eksponensial  e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
```

```
# eksponensial sederhana σ <0 ω = 0
sigma = -1
omega = 0

s = complex(sigma,omega)
sinyal = np.exp(s*t)

# SALIDA - gráfica
plt.figure(3)
plt.plot(t,np.real(sinyal),label='real')
plt.plot(t,np.imag(sinyal),label='imajiner')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('bentuk sinyal eksponensial  e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
```

```
# sinyl  sinus σ=0, ω>0
sigma = 0
omega = 1

s = complex(sigma,omega)
sinyal = np.exp(s*t)

# SALIDA - gráfica
plt.figure(4)
plt.plot(t,np.real(sinyal),label='real')
plt.plot(t,np.imag(sinyal),label='imajiner')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('bentuk sinyal sinus e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
```

```
# sinyal sinus σ=0, ω>0
sigma = 0
omega = 1

s = complex(sigma,omega)
sinyal = np.exp(s*t)

# SALIDA - gráfica
plt.figure(4)
plt.plot(np.real(sinyal),np.imag(sinyal),label='kompleks')
plt.xlabel('real')
plt.ylabel('imajiner')
plt.legend()
plt.margins(0.1)
plt.title('bentuk kompleks sinyal  e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
```
